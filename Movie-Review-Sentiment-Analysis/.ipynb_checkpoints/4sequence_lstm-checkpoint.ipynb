# Represent review with a sequence (w2v1)

```python
num_features1 = 300
length1 = 100
```

```python
# testing case
review_cleaned = train_x_clean[2000]
print len(review_cleaned)
review_cleaned1 = train_x_clean[4000]
print len(review_cleaned1)
```

```python
def text_sequence(words, model, num_features, length=100):
    
    featureVec = np.zeros((length, num_features),dtype="float32")
    vocabulary = set(model.wv.index2word)
    number_of_words = 0
    
    for idx, word in enumerate(words):
        #print word
        if word in vocabulary:
            
            featureVec[number_of_words] = model[word]
            number_of_words += 1
            
            if number_of_words == length:
                break
    
    return featureVec
```

```python
print text_sequence(review_cleaned, model_w2v1, num_features1, length1).shape
```

```python
def dataset_sequence(dataset, model, num_features, length=100):
    new_dataset = []
    for data in dataset:
        new_data = text_sequence(data, model, num_features, length)
        new_dataset.append(new_data)
    return new_dataset
```

```python
def data_generator(features, labels, w2v, num_features, length=100, batch_size=256):
    """
    a batch generator that processes images, create bbox labels,
    and returns (features, labels) based on the mode    
    """
    
    features, labels = shuffle(features, labels, random_state=3)
    sample_size = len(features)
    batch_slices = [slice(i, i + batch_size) for i in range(0, sample_size, batch_size)]
    
    while True:
        for batch in batch_slices:

            batch_new_features = []
            batch_labels=labels[batch]

            for review in features[batch]:

                new_feature = text_sequence(review, w2v, num_features)

                batch_new_features.append(new_feature)

            yield np.array(batch_new_features), np.array(batch_labels)
```

```python
for data in data_generator(train_x_clean, train_y_clean, model_w2v1, 300):
    print data[0].shape
    print len(data[1])
    break
```

```python
validation_x_sequence1 =  dataset_sequence(validation_x_clean, model_w2v1, 300)
test_x_sequence1 =  dataset_sequence(test_x_clean, model_w2v1, 300)
```

# Train model1

```python
def train_w2v1_lstm1():
    
    timesteps = 100
    dimensions = 300
    epochs_number = 40
    #earlyStopping = EarlyStopping(monitor='val_loss', min_delta=0, patience=0, verbose=0, mode='auto')
    
    model = Sequential()
    model.add(LSTM(200,  input_shape=(timesteps, dimensions),  return_sequences=False))
    model.add(Dropout(0.2))
    model.add(Dense(2, input_dim=200, activation='softmax'))
    model.compile(loss="sparse_categorical_crossentropy", optimizer='rmsprop', metrics=["accuracy"])
    
    history = model.fit_generator(generator=data_generator(train_x_clean, train_y_clean, model_w2v1, 300),
                         samples_per_epoch=len(train_x_clean),
                         nb_epoch=epochs_number,
                         validation_data=data_generator(validation_x_clean, validation_y_clean, model_w2v1, 300),
                         nb_val_samples=epochs_number )
   
    return model, history
```

```python
w2v1_lstm1 = train_w2v1_lstm1()
```

```python
model_w2v1_lstm1 = w2v1_lstm1[0]
history_w2v1_lstm1 = w2v1_lstm1[1].history
```

```python
print model_w2v1_lstm1.evaluate(np.array(validation_x_sequence1), np.array(validation_y_clean))
```

```python
model_w2v1_lstm1.save(os.path.join(model_root,'model_w2v1_lstm1'))
with open(os.path.join(model_root,'history_w2v1_lstm1.pkl'), 'wb') as f:
    pickle.dump(history_w2v1_lstm1, f)
```

```python
train_loss = history_w2v1_lstm1['loss']
train_accuracy = history_w2v1_lstm1['acc']
validation_loss = history_w2v1_lstm1['val_loss']
validation_accuracy = history_w2v1_lstm1['val_acc']
epoch = range(40)
```

```python
ax = plt.gca()
ax.plot(epoch, train_loss, label = "Train")
ax.plot(epoch, validation_loss, label = "Validation")
ax.legend()
ax.set_xlabel('Iteration') 
ax.set_ylabel('Loss')
ax.set_title('Model 1: Loss (train vs validation)')
```

```python
ax = plt.gca()
ax.plot(epoch, train_accuracy, label = "Train")
ax.plot(epoch, validation_accuracy, label = "Validation")
ax.legend()
ax.set_xlabel('Iteration') 
ax.set_ylabel('Accuracy')
ax.set_title('Model 1: Accuracy (train vs validation)')
```
